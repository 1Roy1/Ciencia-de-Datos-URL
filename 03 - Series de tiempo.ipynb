```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import yfinance as yf
import numpy as np
```

```python
dta = sm.datasets.co2.load_pandas().data
```

```python
dta.plot()
plt.title('Niveles de CO2')
plt.ylabel('Partes por milln')
plt.xlabel('Ao')
plt.show()
```

```python
ticker_google = 'GOOGL'
start_date = '2025-01-01'
end_date = '2025-12-31'
df_google = yf.download(ticker_google, start = start_date, end=end_date)
```

```python
df_google
```

```python
df_google['Close'].plot()
plt.title('Eje Normal')
plt.show()
```

```python
df_google.info()
```

```python
df_google['LogClose'] = np.log(df_google['Close'])
```
