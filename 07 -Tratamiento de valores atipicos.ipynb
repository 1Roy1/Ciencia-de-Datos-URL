### Puntuacion Z

```python
outliers = []
```

```python
def encontrar_outliers(datos):
    umbral = 3
    media = np.mean(datos)
    desvstd = np.std(datos)
    for xi in datos:
        puntaje_Z = (xi-media) / desvstd
        if(np.abs(puntaje_Z) > umbral):
            outliers.append(xi)
    return media, desvstd, outliers
```

```python
media, desviacion, extremos = encontrar_outliers(muestra)
print(f'La media de la distribucion es:{media:2f}')
print(f'La desviacion de la distribucino es: {desviacion:2f}')
print(f'Los valores extremo de la distribucion son: {extremos}')
```
